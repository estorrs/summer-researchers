## spatial transcriptomics metadata

```python
adata = sc.read_visium('/diskmnt/Projects/Users/estorrs/summer-researchers/data/HT397B1/spatial/spaceranger/HT397B1-S1H2Fs4U1Bp1/')
adata
```

```python
adata.obs
```

```python
# example of an annotation file. In this case we are looking at DCIS and IDC region annotations in this breast cancer sample. I.e. everything that is not a NaN is a spot that is perdominantly tumor
annotations = pd.read_csv(
    '/diskmnt/Projects/Users/estorrs/summer-researchers/data/HT397B1/spatial/annotations/HT397B1-S1H2.csv')
annotations = annotations.set_index('Barcode')
annotations
```

```python
# lets merge it into the adata object, note that the Barcode column of the annotation table matches the spot ids in the adata object
sorted(adata.obs.index)[:5], sorted(annotations.index)[:5]
```

```python
# performing merge
adata.obs = pd.merge(adata.obs, annotations, left_index=True, right_index=True)
adata.obs
```

```python
# visualizing the labels
sc.pl.spatial(adata, color=['celltypespecific'])
```

```python
# we can then make other metadata, like for example labeling all tumor spots
adata.obs['is_tumor'] = ['no' if pd.isnull(x) else 'yes' for x in adata.obs['celltypespecific']]
sc.pl.spatial(adata, color=['is_tumor'])
```
